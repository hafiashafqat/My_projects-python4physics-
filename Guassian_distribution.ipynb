```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
pi = np.pi

def sqrt(x):
    return np.sqrt(x)

def exp(x):
    return np.exp(x)

def f_gaus(x, mu, sig):
    amp = 1.0 / sqrt(2.0 * pi * sig**2)   # amplitude (height)
    arg = (x - mu)**2 / (2.0 * sig**2)    # distance from mean
    return amp * exp(-arg)                # full Gaussian formula
```

```python
m0, dm0 = 1, 0.01  # mean and standard deviation

Npoints = 1000  # number of random data points to generate

m0s = np.random.normal(m0, dm0, Npoints)
```

```python
Nbins = 25  # Number of bins (how detailed the bars will be)

plt.hist(m0s, bins=Nbins)  # Plot the histogram
plt.xlabel('m0')           # Label for x-axis
plt.ylabel('Number of points')  # Label for y-axis
plt.title('Histogram of Gaussian-distributed Data')  # Optional title
plt.show()
```

```python
# Plot the histogram again
plt.hist(m0s, bins=Nbins)

# Define x and tmp for the Gaussian curve
x = np.linspace(m0 - 3*dm0, m0 + 3*dm0, 100) # Generate 100 points between m0 - 3*dm0 and m0 + 3*dm0
tmp = f_gaus(x, m0, dm0) * Npoints * (6*dm0/Nbins) # Scale the Gaussian to match the histogram

# Plot the red Gaussian curve on top
plt.plot(x, tmp, color='r', linewidth=2)

# Labeling
plt.xlabel('m0', size=14)
plt.ylabel('Number of points', size=14)
plt.title('Gaussian Distribution: Histogram + Ideal Curve')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
mu, sigma = 1, 0.01
Npoints = 1000
data = np.random.normal(mu, sigma, Npoints)
Nbins = 25
hist, bins, _ = plt.hist(data, bins=Nbins, alpha=0.6, color='skyblue', edgecolor='black')
x = np.arange(mu - 4*sigma, mu + 4*sigma, sigma / 100.)
def gaussian(x, mu, sigma):
    amp = 1.0 / np.sqrt(2 * np.pi * sigma**2)
    exponent = -((x - mu)**2) / (2 * sigma**2)
    return amp * np.exp(exponent)
curve = gaussian(x, mu, sigma)
curve = curve * max(hist) / max(curve)
plt.plot(x, curve, color='red', linewidth=2)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram with Gaussian Curve')
plt.show()
```
