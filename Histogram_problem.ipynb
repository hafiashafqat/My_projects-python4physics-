Histogram of Measurement Noise

You are simulating an experiment where you measure the length of a metal rod 1000 times. The true length is 50 cm, but your instrument introduces random Gaussian noise with a standard deviation of 2 cm.

```python
import numpy as np
import matplotlib.pyplot as plt
mu, sigma = 50, 2
Npoints = 1000
data = np.random.normal(mu, sigma, Npoints)
Nbins = 25
hist, bins, _ = plt.hist(data, bins=Nbins, alpha=0.6, color='purple', edgecolor='white')
x = np.arange(mu - 4*sigma, mu + 4*sigma, sigma / 100.)
def gaussian(x, mu, sigma):
    amp = 1.0 / np.sqrt(2 * np.pi * sigma**2)
    exponent = -((x - mu)**2) / (2 * sigma**2)
    return amp * np.exp(exponent)
curve = gaussian(x, mu, sigma)
curve = curve * max(hist) / max(curve)
plt.plot(x, curve, color='red', linewidth=2)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram with Gaussian Curve')
plt.show()
```

You are analyzing measurements from a particle detector. There’s a known Gaussian-shaped signal (a bump) that represents a rare particle event. But the data is full of noise. Can you recover the signal?

```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Create x-axis (indexes from 0 to 999)
x = np.arange(0, 1000)

# Step 2: Define the Gaussian function
def gaussian(x, mu, sigma, A):
    exponent = -((x - mu)**2) / (2 * sigma**2)
    return A * np.exp(exponent)

# Step 3: Define parameters for the bump
mu = 500        # center of the bump
sigma = 20      # standard deviation
A = 5           # amplitude of the bump

# Step 4: Generate the Gaussian signal
signal = gaussian(x, mu, sigma, A)

# Step 5: Plot the signal
plt.plot(x, signal, color='red', linestyle='--', label='Gaussian Signal')
plt.xlabel('Index (x)')
plt.ylabel('Amplitude')
plt.title('Gaussian Bump Signal (Centered at x = 500)')
plt.legend()
plt.grid(True)
plt.show()
```
