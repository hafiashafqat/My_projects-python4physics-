```python
import numpy as np
import matplotlib.pyplot as plt

def calcpi(Nt):
    x0, y0 = np.random.random(Nt), np.random.random(Nt)
    x = x0 * 2.0 - 1.0
    y = y0 * 2.0 - 1.0
    r = np.sqrt(x**2 + y**2)

    # Boolean arrays for inside and outside the circle
    inside = r < 1
    outside = r >= 1

    # Separate points
    xin = x[inside]
    yin = y[inside]
    xout = x[outside]
    yout = y[outside]

    # Estimate pi
    pi_estimate = 4.0 * len(xin) / Nt
    return xin, yin, xout, yout, pi_estimate

# Run the experiment
Nt = 1000  # Try 1000 random points
xin, yin, xout, yout, pi_ran = calcpi(Nt)

# Plot the result
plt.figure(figsize=(6,6))
plt.scatter(xin, yin, color='blue', label='Inside Circle', s=10)
plt.scatter(xout, yout, color='red', label='Outside Circle', s=10)
circle = plt.Circle((0, 0), 1, color='black', fill=False, linewidth=2)
plt.gca().add_patch(circle)
plt.xlabel("x")
plt.ylabel("y")
plt.title(f"Monte Carlo Pi Estimation\nEstimated π ≈ {pi_ran:.5f}")
plt.legend()
plt.axis('equal')
plt.grid(True)
plt.show()
```
